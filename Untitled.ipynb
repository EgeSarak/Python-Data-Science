```python
def PrimeTime(num):
    test=0
    if num %2==0:
        print("False")
    else:
        sqrt=int(num**(0.5))


    for i in range(2, sqrt+1):    
        if num % i == 0:
            test += 1       
        if test == 0:
            print("true")
        else:
            print("false")
            
print(PrimeTime(2))    
```
