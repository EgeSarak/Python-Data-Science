## MAtplotlib example workflow

```python
#1. import matplotlib and get it ready for plotting 
%matplotlib inline
import matplotlib.pyplot as plt

#2. prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#3 setup plot
fig, ax=plt.subplots(figsize=(10,10)) # (width,height)

#4 plot data
ax.plot(x,y)

#5 customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#6. Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

```python
over_50=df3[df3["age"]> 50]
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));
```

```python
# Object Oriented Method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

#ax.set_xlim([45,100]);
```

```python
#Object Oriented Method

# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"]);

#cutomize the plot

ax.set(title="Heart Disease an Cholesterol levels",
      xlabel="Age",
      ylabel="cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                    c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Choloesterol levels",
       xlabel="Age",
       ylabel="Cholesterol");

#add a legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line(meanline)
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--");


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

#customize ax1
ax1.set(title="heart diesease and max heart rate",
       xlabel="Age",
       ylabel="max heart rate");

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                    c=over_50["target"])

#customize ax0
ax0.set(title="Heart Disease and Choloesterol levels",
       ylabel="Cholesterol");

#add a legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line(meanline)
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--");


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

#customize ax1
ax1.set(title="heart diesease and max heart rate",
       xlabel="Age",
       ylabel="max heart rate");

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart  Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#set the style

plt.style.use("seaborn-whitegrid")

# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

# Plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                  #cmap="winter",
                  #cmap="plasma",
                  cmap="summer"); # this changes the colour schema

#customize the plot

ax.set(title="Heart Disease an Cholesterol levels",
      xlabel="Age",
      ylabel="cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#customizing the y and x axis limitations

# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10),
                          sharex=True)

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                    c=over_50["target"],
                   cmap="winter");

#customize ax0
ax0.set(title="Heart Disease and Choloesterol levels",
       ylabel="Cholesterol");

#change the x axis limits
ax0.set_xlim([50,80])

#add a legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line(meanline)
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--");


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter");

#customize ax1
ax1.set(title="heart diesease and max heart rate",
       xlabel="Age",
       ylabel="max heart rate");

#change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart  Disease Analysis",fontsize=16,fontweight="bold");
```
