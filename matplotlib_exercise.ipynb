```python
favori_yemek_ucret={"Almond butter":10,
                   "Blueberries":5,
                   "Eggs":6}
```

```python
fig,ax = plt.subplots()
ax.bar(favori_yemek_ucret.keys(),favori_yemek_ucret.values())

ax.set(title="favori yemek cretleri",
      xlabel="Food",
      ylabel="Price");
```

```python
fig,ax = plt.subplots()
ax.barh(list(favori_yemek_ucret.keys()),
        list(favori_yemek_ucret.values()))

ax.set(title="favori yemek cretleri",
      xlabel="Food",
      ylabel="Price");
```

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace("[\$\,\.]","")
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("13/1/2020",periods=len(car_sales))
car_sales
```

```python
car_sales.plot(x="Sale Date",y= "Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales.plot(x="Odometer (KM)", y ="Price",kind="scatter");
```
